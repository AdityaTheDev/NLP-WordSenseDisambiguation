```python
import csv
  
with open('corpus.csv', 'r') as read_obj:
  
    csv_reader = csv.reader(read_obj)
  
    corpus = list(csv_reader)
  


```

```python
with open('query.csv', 'r') as read_obj:
  
    csv_reader = csv.reader(read_obj)
  
    q = list(csv_reader)
```

```python
total=len(corpus)
total
```

```python
V=int(input("Enter the value of V :"))
```

```python
def condition_p(w1,w2,corpus):
    cnt=0
    for document in corpus:
        if(document[len(document)-1]==w2):
            for w in document:
                if(w==w1):
                    cnt=cnt+1
                        
    return cnt                   
```

```python
def class_p(w,corpus):
    cnt=0
    for document in corpus:
        if(document[len(document)-1]==w):
            cnt=cnt+1
            
    return cnt         
```

```python
condition_p("head","furniture",corpus)
```

```python
import math
```

```python
def final_score(sense,q,corpus):
    ans=math.log2(class_p(sense,corpus)/total)
    for smallist in q:
        for word in smallist:
            ans=ans+math.log2(((condition_p(word,sense,corpus)+1)/(class_p(sense,corpus)+V)))
        
    return ans                     
                                               
```

```python
f=final_score("furniture",q,corpus)
f
```

```python
p=final_score("position",q,corpus)
p
```

```python
if(p>f):
    print("The query belongs to position")
else:
    print("The query belongs to furniture")
```
